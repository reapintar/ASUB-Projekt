## Brain tumor classification using MRI images

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import cv2
from sklearn.utils import shuffle
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
from collections import Counter
from tqdm import tqdm
from sklearn.model_selection import train_test_split
```

```python
dataDir = "Data"

labels = []
for dirname, dirnames, filenames in os.walk(os.path.join(dataDir, "Training")):
  if(len(dirnames) > 0):
    labels = dirnames
    break
print(labels)
```

```python
image_size = 112
X_train = []
y_train = []
for i in labels:
    folderPath = os.path.join('Data','Training',i)
    for j in tqdm(os.listdir(folderPath)):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.resize(img,(image_size, image_size))
        img = cv2.normalize(img, img, 0, 255, cv2.NORM_MINMAX)
        img = cv2.medianBlur(img, 5)
        X_train.append(img)
        y_train.append(labels.index(i))
        
for i in labels:
    folderPath = os.path.join('Data','Testing',i)
    for j in tqdm(os.listdir(folderPath)):
        img = cv2.imread(os.path.join(folderPath,j))
        img = cv2.resize(img,(image_size,image_size))
        img = cv2.normalize(img, img, 0, 255, cv2.NORM_MINMAX)
        img = cv2.medianBlur(img, 5)
        X_train.append(img)
        y_train.append(labels.index(i))
        
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
c = Counter(y_train)
frekvencije = [c[i] for i in range(0, len(labels))]
#print(c)
tumor = ["bez tumora","meningeomi","gliomi","tumor hipofize"]
fig, ax = plt.subplots(figsize = (10, 7))
bars = plt.bar(tumor, frekvencije, color = ["royalblue", "lightgray", "lightblue","coral"])
ax.bar_label(bars,size = 12)

for item in ([ ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(12)
plt.xlabel("Vrsta tumora")
plt.ylabel("Broj slika")
plt.title("Distribucija tumora")
ax.title.set_fontsize(15)
plt.show()
```

```python
X_train, y_train = shuffle(X_train,y_train, random_state=152)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_train,y_train, test_size=0.2,random_state=156)
print(f"velicina skupa za treniranje {len(X_train)}")
print(y_train)
print(f"velicina skupa za testiranje: {len(X_test)}")
plt.imshow(X_train[0])
```

```python
c = Counter(y_train)
frekvencije = [c[i] for i in range(0, len(labels))]
#print(c)
tumor = ["bez tumora","meningeomi","gliomi","tumor hipofize"]
fig, ax = plt.subplots(figsize = (10, 7))
bars = plt.bar(tumor, frekvencije, color = ["royalblue", "lightgray", "lightblue","coral"])
ax.bar_label(bars,size = 12)

for item in ([ ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(12)
plt.xlabel("Vrsta tumora")
plt.ylabel("Broj slika")
plt.title("Distribucija tumora u skupu za treniranje")
ax.title.set_fontsize(15)
plt.show()
```

```python
c = Counter(y_test)
frekvencije = [c[i] for i in range(0, len(labels))]
#print(c)
tumor = ["bez tumora","meningeomi","gliomi","tumor hipofize"]
fig, ax = plt.subplots(figsize = (10, 7))
bars = plt.bar(tumor, frekvencije, color = ["royalblue", "lightgray", "lightblue","coral"])
ax.bar_label(bars,size = 12)

for item in ([ ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(12)
plt.xlabel("Vrsta tumora")
plt.ylabel("Broj slika")
plt.title("Distribucija tumora u skupu za testiranje")
ax.title.set_fontsize(15)
plt.show()
```

```python
y_train = tf.keras.utils.to_categorical(y_train)
```

```python
y_test = tf.keras.utils.to_categorical(y_test)
print(y_train)
```

```python
from tensorflow.keras.applications import EfficientNetV2L
efficient_net = EfficientNetV2L(weights='imagenet', include_top=False, input_shape=(image_size, image_size, 3))
```

```python
model = efficient_net.output
model = tf.keras.layers.GlobalAveragePooling2D()(model)
model = tf.keras.layers.Dropout(0.5)(model)
model = tf.keras.layers.Dense(4, activation='softmax')(model)
model = tf.keras.models.Model(inputs=efficient_net.input, outputs=model)
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpoint = tf.keras.callbacks.ModelCheckpoint("model.h5", monitor='val_accuracy', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', save_freq='epoch')
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=2, min_delta=0.001)
```

```python
history = model.fit(tf.convert_to_tensor(X_train), tf.convert_to_tensor(y_train), epochs=10, batch_size=32, validation_split=0.1, callbacks=[checkpoint, reduce_lr])
```

```python
prediction = model.predict(tf.convert_to_tensor(X_test))
```

```python
prediction_y = []
for row in prediction:
  index = np.argmax(row)
  prediction_y.append(index)

print(prediction_y)
y_test_numerical = np.argmax(y_test, axis=1)
print(y_test_numerical)
```

```python
print(classification_report(y_test_numerical,prediction_y))
```

```python
print(confusion_matrix(y_test_numerical,prediction_y))
```

```python
fig,ax=plt.subplots(1,1,figsize=(14,7))
colors_green = ['#01411C','#4B6F44','#4F7942','#74C365','#D0F0C0']
colors_dark = ["#1F1F1F", "#313131", '#636363', '#AEAEAE', '#DADADA']

import seaborn as sns
sns.heatmap(confusion_matrix(y_test_numerical,prediction_y),ax=ax,xticklabels=labels,yticklabels=labels,annot=True,
           cmap=colors_green[::-1],alpha=0.7,linewidths=2,linecolor=colors_dark[3],fmt='d')
fig.text(s='Heatmap of the Confusion Matrix',size=18,fontweight='bold',
             fontname='monospace',color=colors_dark[1],y=0.92,x=0.28,alpha=0.8)

plt.show()
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]+0.5, str(y[i])+"%", ha = 'center',size = 12)
        
tocnosti = [83.8,97.9,93.3]
modeli = ["SVM","EfficientNetV2L","GoogLeNet"]
fig, ax = plt.subplots(figsize = (7, 7))
bars = plt.bar(modeli, tocnosti, color = [ "lightgray", "lightblue","coral"])


for item in ([ ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(12)
addlabels(modeli,tocnosti)
plt.xlabel("Klasifikator")
plt.ylabel("Točnost (%)")
plt.title("Točnost klasifikatora na skupu za ispitivanje")
ax.title.set_fontsize(15)
plt.show()
```
